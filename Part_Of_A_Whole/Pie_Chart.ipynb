```python
import matplotlib.pyplot as plt
```

```python
# 색상
colors = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897']

plt.rcParams["figure.figsize"] = (20,5)
names='groupA', 'groupB', 'groupC', 'groupD'
size_of_groups=[12,11,3,30]

plt.pie(size_of_groups, 
        labels= names, 
        labeldistance = 1.15,
        wedgeprops = {'linewidth' : 3, 
                      'edgecolor' : 'white'},
        colors=colors)
plt.show()
```
