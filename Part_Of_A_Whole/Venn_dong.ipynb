```python
# 특정 레이블 사용자 정의
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

v = venn3(subsets = (10, 8, 22, 6,9,4,2),                   # 값 지정
          set_labels = ('group A', 'Group B', 'Group C'),   # 레이블 지정
          )
v.get_label_by_id('A').set_text('My Favourite group!')      # 특정 레이블 사용자 정의

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3, venn3_circles

v = venn3(subsets = (10, 8, 22, 6,9,4,2),                   # 값 지정
          set_labels = ('group A', 'Group B', 'Group C'),   # 레이블 지정
         )
# 원에 대한 설정
c = venn3_circles(subsets= (10, 8, 22, 6,9,4,2),
                  linestyle='dashed',               # 원 테두리 스타일
                  linewidth=1,                      # 원 테두리 굵기
                  color="grey"                      # 원 테두리 색상
                  )

plt.show()
```

```python
# 특정 그룹 설정
import matplotlib.pyplot as plt
from matplotlib_venn import venn3, venn3_circles

v = venn3(subsets = (10, 8, 22, 6,9,4,2),                   # 값 지정
          set_labels = ('group A', 'Group B', 'Group C'),   # 레이블 지정
         )
# 원에 대한 설정
c = venn3_circles(subsets= (10, 8, 22, 6,9,4,2),
                  linestyle='dashed',               # 원 테두리 스타일
                  linewidth=1,                      # 원 테두리 굵기
                  color="grey"                      # 원 테두리 색상
                  )

# 첫번째 그룹에 대한 설정
c[0].set_lw(8.0)                # 선 너비
c[0].set_ls('dotted')           # 선 스타일
c[0].set_color('skyblue')       # 색상

plt.show()
```

```python
# 배경색 설정
from matplotlib import pyplot as pyplot
from matplotlib_venn import venn2

v = venn2( (10, 20, 10), alpha = 1)

# 배경 설정
plt.gca().set_facecolor('skyblue')      # 배경색
plt.gca().set_axis_on()                 # 배경색 적용

plt.show()
```

```python
# 정교한 벤다이어그램
from matplotlib import pyplot as plt
import numpy as np
from matplotlib_venn import venn3, venn3_circles

v = venn3(subsets = (1, 1, 1, 1, 1, 1, 1),
          set_labels = ('A', 'B', 'C')
          )

v.get_patch_by_id('100').set_alpha(1.0)                                 # 첫번째 원 투명도
v.get_patch_by_id('100').set_color('white')                             # 첫번째 원 색상
v.get_label_by_id('100').set_text('Unknown')                            # 첫번째 원 값
v.get_label_by_id('A').set_text('Set "A"')
c = venn3_circles(subsets=(1, 1, 1, 1, 1, 1, 1), linestyle='dashed')    # 전체 원 설정
c[0].set_lw(1.0)            # 첫번쨰 원 테투리 너비
c[0].set_ls('dotted')       # 첫번째 원 테투리 스타일

plt.title("Sample Venn diagram")

# 주석 달기
plt.annotate('Unknown set',
             xy=v.get_label_by_id('100').get_position() - np.array([0, 0.05]),
             xytext=(-70,-70),
             ha='center',
             textcoords='offset points',
             bbox=dict(boxstyle='round,pad=0.5', fc='gray', alpha=0.1),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.5',color='gray')
             )

plt.show()
```
