```python
# 도넛 차트
import matplotlib.pyplot as plt

size_of_groups=[12, 11, 3, 30]

plt.pie(size_of_groups)                 # 파이차트

my_circle = plt.Circle((0, 0),          # 원 중심 위치
                       0.7,             # 원 크기
                       color='white'    # 원 색상
                       )

p = plt.gcf()                           # 현재 figure   반환
p.gca().add_artist(my_circle)           # 추가 반영

plt.show()
```

```python
# 색상지정
import matplotlib.pyplot as plt

names = ['groupA', 'groupB', 'groupC', 'groupD']
size=[12, 11, 3, 30]

# 중심원
my_circle = plt.Circle((0, 0),          # 원 중심 위치
                       0.7,             # 원 크기
                       color='white'    # 원 색상
                       )

# 파이차트
plt.pie(size,                                       # 값
        labels=names,                               # 레이블
        colors=['red', 'green', 'blue', 'skyblue']  # 색상 지정
        )

p = plt.gcf()                           # 현재 figure   반환
p.gca().add_artist(my_circle)           # 추가 반영

plt.show()
```

```python
# 색상지정: 팔레트 색상 사용
#!pip install palettable
import matplotlib.pyplot as plt
from palettable.colorbrewer.qualitative import Pastel1_7

names = ['groupA', 'groupB', 'groupC', 'groupD']
size = [12,11,3,30]

#중심원
my_circle = plt.Circle((0,0), 0.7, color='white')

plt.pie(size,
        labels=names,
        colors=Pastel1_7.hex_colors
        )

p = plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```

```python
# 레이블 표시
import matplotlib.pyplot as plt

names = ['groupA', 'groupB', 'groupC', 'groupD']
size = [12,11,3,30]

#중심원
my_circle = plt.Circle((0,0), 0.7, color='white')

plt.pie(size,
        labels=names,
        labeldistance=0.45          # 레이블을 표시할 거리
        )

p = plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```

```python
# 레이블 색상 지정
import matplotlib.pyplot as plt

names = ['groupA', 'groupB', 'groupC', 'groupD']
size = [12,11,3,30]

#중심원
my_circle = plt.Circle((0,0), 0.7, color='white')

plt.rcParams['text.color'] = 'red'          # 레이블 색상 지정
plt.pie(size,
        labels=names,
        labeldistance=0.5         # 레이블을 표시할 거리
        )

p = plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```

```python
# 경계선의 너비 지정
import matplotlib.pyplot as plt

names = ['groupA', 'groupB', 'groupC', 'groupD']
size = [12,11,3,30]

#중심원
my_circle = plt.Circle((0,0), 0.7, color='white')

plt.rcParams['text.color'] = 'red'          # 레이블 색상 지정
plt.pie(size,
        labels=names,
        labeldistance=0.5,        # 레이블을 표시할 거리
        wedgeprops = {'linewidth':5, 'edgecolor':'black'}           # 경계선 너비 & 색상
        )

p = plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```

```python
# 배경색 지정
import matplotlib.pyplot as plt

names = ['groupA', 'groupB', 'groupC', 'groupD']
size = [12,11,3,30]

fig = plt.figure()
fig.patch.set_facecolor('black')            # 배경색 지정

#중심원
my_circle = plt.Circle((0,0), 0.7, color='black')

plt.rcParams['text.color'] = 'white'          # 레이블 색상 지정
plt.pie(size,
        labels=names,
        #labeldistance=0.5,        # 레이블을 표시할 거리
        wedgeprops = {'linewidth':5, 'edgecolor':'black'}           # 경계선 너비 & 색상
        )

p = plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```
