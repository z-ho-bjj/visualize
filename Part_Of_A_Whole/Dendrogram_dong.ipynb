```python
# 클러스터링을 위한 덴드로그램
import pandas as pd
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np

url = 'https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/mtcars.csv'
df = pd.read_csv(url)
df = df.set_index('model')

Z = linkage(df, 'ward')         # 계층적 군집 계산

dendrogram(Z,
           leaf_rotation=90,            # 리프 레이블 회전각
           leaf_font_size=8,            # 리프 레이블 폰트 크기
           labels=df.index              # 레이블 지정
           )

plt.show()
```

```python
# 클러스터 수 지정
import pandas as pd
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
 
url = 'https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/mtcars.csv'
df = pd.read_csv(url)
df = df.set_index('model')
 
Z = linkage(df, 'ward')
 
dendrogram(Z,
           color_threshold=240              # 클러스터 경계값
           )

plt.axhline(y=240,                          # 경계선 값
            c='grey',                       # 경계선 색상
            lw=1,                           # 경계선 굵기
            linestyle='dashed'              # 경계선 스타일
            )

plt.show()
```

```python
# 클러스터 노드 색상
import pandas as pd
from matplotlib import pyplot as plt
from scipy.cluster import hierarchy
import numpy as np

url = 'https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/mtcars.csv'
df = pd.read_csv(url)
df = df.set_index('model')

Z = hierarchy.linkage(df, 'ward')

hierarchy.set_link_color_palette(['#b30000','#996600', '#b30086'])          # 색상 지정

hierarchy.dendrogram(Z,
                     color_threshold=240,                   # 클러스터 경계값
                     above_threshold_color='grey'           # 클러스터 외 노드 색상
                     )

plt.axhline(y=240,                          # 경계선 값
            c='grey',                       # 경계선 색상
            lw=1,                           # 경계선 굵기
            linestyle='dashed'              # 경계선 스타일
            )

plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
from scipy.cluster import hierarchy
import numpy as np

url = 'https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/mtcars.csv'
df = pd.read_csv(url)
df = df.set_index('model')

Z = hierarchy.linkage(df, 'ward')

hierarchy.dendrogram(Z,
                     truncate_mode = 'lastp',           # 맨 아이잎 수 설정
                     p=6                                # 맨 아랫잎 수
                     )
plt.show()

hierarchy.dendrogram(Z,
                     truncate_mode = 'level',           # 트리의 'p'수준 이하
                     p=2                                # 수준 수
                     )
plt.show()
```

```python
# 덴드로그램 방향 설정
import pandas as pd
from matplotlib import pyplot as plt
from scipy.cluster import hierarchy
import numpy as np

url = 'https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/mtcars.csv'
df = pd.read_csv(url)
df = df.set_index('model')

Z = hierarchy.linkage(df, 'ward')

hierarchy.dendrogram(Z,
                     orientation="right",           # 방향 설정
                     labels=df.index                # 레이블 설정
                     )
plt.show()

hierarchy.dendrogram(Z,
                     orientation="left",            # 방향 설정
                     labels=df.index                # 레이블 설정
                     )
plt.show()
```
