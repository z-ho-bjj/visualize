```python
# 파이차트
import matplotlib.pyplot as plt

size_of_groups = [12, 11, 3, 30]

plt.pie(size_of_groups)
plt.show()
```

```python
# 기본 파이차트
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (20, 5)           # 크기 설정

values = [12, 11, 3, 30]

plt.pie(values)
plt.show()
```

```python
# 레이블 추가
import matplotlib.pyplot as plt

names = ['groupA', 'groupB', 'groupC', 'groupD']
values=[12, 11, 3, 30]

plt.pie(values,
        labels=names,           # 레이블 표시
        labeldistance=1.15      # 레이블 위치 지정
        )

plt.show()
```

```python
# 경계선 간격 설정
import matplotlib.pyplot as plt

names = ['groupA', 'groupB', 'groupC', 'groupD']
values=[12, 11, 3, 30]

plt.pie(values,
        labels=names,           # 레이블 표시
        labeldistance=1.15,     # 레이블 위치 지정
        wedgeprops = {'linewidth':3, 'edgecolor':'white'}       # 경계선 굵기 & 색상
        )

plt.show()
```

```python
# 색상 설정
import matplotlib.pyplot as plt

names = ['groupA', 'groupB', 'groupC', 'groupD']
values=[12, 11, 3, 30]
colors = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897']           # 색상 지정

plt.pie(values,
        labels=names,           # 레이블 표시
        labeldistance=1.15,     # 레이블 위치 지정
        wedgeprops = {'linewidth':3, 'edgecolor':'white'},      # 경계선 굵기 & 색상
        colors=colors           # 색상 적용
        )

plt.show()
```
