```python
#   
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set(style="darkgrid")
df = sns.load_dataset('iris')

ax = sns.violinplot(x="species", y="sepal_length", data=df)

medians = df.groupby(['species'])['sepal_length'].median().values
nobs = df['species'].value_counts().values   # [50 50 50]
nobs = [str(x) for x in nobs.tolist()]       # ['50', '50', '50']
nobs = ["n: " + i for i in nobs]             # ['n: 50', 'n: 50', 'n: 50']

pos = range(len(nobs))
for tick, label in zip(pos, ax.get_xticklabels()):
    ax.text(pos[tick], medians[tick] + 0.03, nobs[tick],
            horizontalalignment='center',
            size='small',
            color='w',
            weight='semibold')
plt.show()
```

```python
### 
import seaborn as sns
import matplotlib.pyplot as plt

# data
df = sns.load_dataset('iris')

# 
sns.set(style="darkgrid")

#    
my_pal = {species: "r" if species == "virginica" else "b" for species in df["species"].unique()}

#   
my_order = df.groupby(by=["species"])["sepal_length"].median().iloc[::-1].index

# violinplot
ax = sns.violinplot( x="species",
                     y="sepal_length",
                     data=df,
                     linewidth=2,          # 
                     width=0.8,            #   
                     palette=my_pal,       #  ( )
                     #color="red"           #  
                     order=my_order        #  
                     )

#  
medians = df.groupby(['species'])['sepal_length'].median().iloc[::-1].values
nobs = df['species'].value_counts().values   # [50 50 50]
nobs = [str(x) for x in nobs.tolist()]       # ['50', '50', '50']
nobs = ["n: " + i for i in nobs]             # ['n: 50', 'n: 50', 'n: 50']

pos = range(len(nobs))
for tick, label in zip(pos, ax.get_xticklabels()):
    ax.text(pos[tick], medians[tick] + 0.03, nobs[tick],
            horizontalalignment='center',
            size='small',
            color='w',
            weight='semibold')

plt.show()
```
