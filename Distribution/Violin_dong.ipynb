```python
# 목록을 직접 지정
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="darkgrid")
df = sns.load_dataset('iris')

sns.violinplot(x='species', y='sepal_length', data=df, order=["versicolor", "virginica", "setosa"])
plt.show()
```

```python
# 중앙값 내림차순 정렬
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="darkgrid")
df = sns.load_dataset('iris')

my_order = df.groupby(by=["species"])["sepal_length"].median().iloc[::-1].index

sns.violinplot(x='species', y='sepal_length', data=df, order=my_order)
plt.show()
```

```python
# 그룹별 바이올린플롯 'hue'
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="darkgrid")
df = sns.load_dataset('tips')

sns.violinplot(x="day", y="total_bill", hue="smoker", data=df, palette="Pastel1")
plt.show()
```

```python
# 관측치 수를 표시
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set(style="darkgrid")
df = sns.load_dataset('iris')

ax = sns.violinplot(x="species", y="sepal_length", data=df)

medians = df.groupby(['species'])['sepal_length'].median().values
nobs = df['species'].value_counts().values   # [50 50 50]
nobs = [str(x) for x in nobs.tolist()]       # ['50', '50', '50']
nobs = ["n: " + i for i in nobs]             # ['n: 50', 'n: 50', 'n: 50']

pos = range(len(nobs))
for tick, label in zip(pos, ax.get_xticklabels()):
    ax.text(pos[tick], medians[tick] + 0.03, nobs[tick],
            horizontalalignment='center',
            size='small',
            color='w',
            weight='semibold')
plt.show()
```

```python
### 정리
import seaborn as sns
import matplotlib.pyplot as plt

# data
df = sns.load_dataset('iris')

# 배경
sns.set(style="darkgrid")

# 그룹 강조 색상 지정
my_pal = {species: "r" if species == "virginica" else "b" for species in df["species"].unique()}

# 중앙값 내림차순 정렬
my_order = df.groupby(by=["species"])["sepal_length"].median().iloc[::-1].index

# violinplot
ax = sns.violinplot( x="species",
                     y="sepal_length",
                     data=df,
                     linewidth=2,          # 윤곽강조
                     width=0.8,            # 그래프 폭 조정
                     palette=my_pal,       # 변수별 색상(그룹강조할때 용이)
                     #color="red"           # 균일한 색상
                     order=my_order        # 그래프 순서
                     )

# 관측치 표시
medians = df.groupby(['species'])['sepal_length'].median().iloc[::-1].values
nobs = df['species'].value_counts().values   # [50 50 50]
nobs = [str(x) for x in nobs.tolist()]       # ['50', '50', '50']
nobs = ["n: " + i for i in nobs]             # ['n: 50', 'n: 50', 'n: 50']

pos = range(len(nobs))
for tick, label in zip(pos, ax.get_xticklabels()):
    ax.text(pos[tick], medians[tick] + 0.03, nobs[tick],
            horizontalalignment='center',
            size='small',
            color='w',
            weight='semibold')

plt.show()
```
