```python
# bin 설정
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="darkgrid")
df = sns.load_dataset("iris")

sns.histplot(data=df,
             x="sepal_length",
             kde=True,              # kde 곡선
             bins=20                # 막대 수
             )
plt.show()
```

```python
# 상자그림
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="darkgrid")
df = sns.load_dataset("iris")

f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})

# 상자그림
sns.boxplot(x="sepal_length",
            data=df,
            ax=ax_box
            )
# 히스토그램
sns.histplot(data=df,
             x="sepal_length",
             kde=True,              # kde 곡선
             bins=20,               # 막대 수
             ax=ax_hist             # 축설정
             )

ax_box.set(xlabel='')   #상자그림에 x라벨 제거
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="darkgrid")
df = sns.load_dataset("iris")

fig, axs = plt.subplots(2, 2, figsize=(7,7))

sns.histplot(data=df,
             x="sepal_length",
             kde=True,
             color="skyblue",
             ax=axs[0, 0]
             )
sns.histplot(data=df,
             x="sepal_width",
             kde=True,
             color="olive",
             ax=axs[0, 1]
             )
sns.histplot(data=df,
             x="petal_length",
             kde=True,
             color="gold",
             ax=axs[1, 0]
             )
sns.histplot(data=df,
             x="petal_width",
             kde=True,
             color="teal",
             ax=axs[1, 1]
             )

plt.show()
```

```python
# jointplot
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

sns.jointplot(x=df["sepal_length"],
              y=df["sepal_width"],
              kind='scatter'
              )
sns.jointplot(x=df["sepal_length"],
              y=df["sepal_width"],
              kind='hex'
              )
sns.jointplot(x=df["sepal_length"],
              y=df["sepal_width"],
              kind='kde'
              )

plt.show()
```

```python
sns.set(style="white", color_codes=True)

sns.jointplot(x=df["sepal_length"],
              y=df["sepal_width"],
              kind='scatter',
              s=200,                    # scatter size
              color='r',                # 색상
              edgecolor="skyblue",      # 테두리 색상
              linewidth=2               # 테두리 굵기
              )
sns.jointplot(x=df["sepal_length"],
              y=df["sepal_width"],
              kind='kde',
              color="skyblue"
              )

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

sns.jointplot(x=df["sepal_length"],
              y=df["sepal_width"],
              kind='hex',             # 종류 { “scatter” | “kde” | “hist” | “hex” | “reg” | “resid” }
              marginal_kws=dict(bins=30, fill=True, kde=True)    # 추가 키워드 인수
              )

plt.show()
```

```python
# 비율 설정
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

sns.jointplot(x=df["sepal_length"],
              y=df["sepal_width"],
              kind='kde',             # 종류 { “scatter” | “kde” | “hist” | “hex” | “reg” | “resid” }
              color='red',
              space=0                 # 축 사이의 공간
              )
sns.jointplot(x=df["sepal_length"],
              y=df["sepal_width"],
              kind='kde',             # 종류 { “scatter” | “kde” | “hist” | “hex” | “reg” | “resid” }
              color='blue',
              space=3                 # 축 사이의 공간
              )
sns.jointplot(x=df["sepal_length"],
              y=df["sepal_width"],
              kind='kde',             # 종류 { “scatter” | “kde” | “hist” | “hex” | “reg” | “resid” }
              color='grey',
              ratio=1                 # 높이 비율
              )
plt.show()
```
