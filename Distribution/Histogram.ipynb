```python
# library & dataset
import seaborn as sns
df = sns.load_dataset('iris')
```

```python
# distplot은 이제 사라질 예정이라 histplot으로 진행
sns.set_style('whitegrid')
sns.histplot(data=df['sepal_length'], kde=True, bins = 20, color='r')
```

```python
# element를 step으로 바꾸면 bar들을 붙일 수 있다.
sns.histplot(data=df['sepal_length'], kde=True, bins = 20, color='g', element='step')
```

```python
# FacetGrid 이용!
# matplotlib의 subplots를 사용하지 않고 여러개 그리기
# col_wrap의 수로 컬럼수를 조절하여 카테고리별로 나타낸다.
g = sns.FacetGrid(df, col='species', hue='species', col_wrap=2)

g = g.map(sns.histplot,"sepal_length", fill=True, common_norm=False, alpha=1, legend=False)

# title 지정
g = g.set_titles("{col_name}")
import matplotlib.pyplot as plt
plt.show()
```

```python
fig, ax = plt.subplots(2,2)
fig.set_size_inches((8,8))
multiple_list = ['layer', 'dodge', 'stack', 'fill']
for i in range(4):
    row = i//2
    col = i%2
    cur_ax = ax[row, col]
    sns.histplot(data=df, x='sepal_length',
                      hue='species',
                      hue_order=['setosa', 'versicolor','virginica'],
                      palette='cool', ## 컬러맵 지정
                      multiple=multiple_list[i], ## 여러 히스토그램 그리는 방식
                      ax=cur_ax,
                     ) 
    cur_ax.set_title(multiple_list[i])
plt.tight_layout()
```
