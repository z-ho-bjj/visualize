```python
import seaborn as sns
import pandas as pd
import numpy as np
np.random.seed(0)

df = pd.DataFrame(np.random.random((100, 5)), columns=["a", "b", "c", "d", "e"])

# 상관관계 행렬 생성
corr_matrix = df.corr()

# 상관관계 히트맵을 위한 마스크 생성
# triu_indices_from 함수는 상삼각 행렬(upper triangle matrix)의 인덱스를 반환, 삼각 행렬 아래 부분을 가리는 역할
mask = np.zeros_like(corr_matrix)
mask[np.triu_indices_from(mask)] = True

# 상관관계 히트맵 그리기
sns.heatmap(corr_matrix, mask=mask, square=True)
```

```python
# 연속형 데이터 불연속으로 변환 후 히트맵 표현
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(6, 6))

df_q = pd.DataFrame()
for col in df:
    df_q[col] = pd.to_numeric(pd.qcut(df[col], 3, labels=list(range(3))))

sns.heatmap(df_q)
plt.show()
```

```python
# 열 정규화
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 음수기호 표시
plt.rcParams['axes.unicode_minus'] = False

df = pd.DataFrame(np.random.randn(10,10)*4+3)
df[1]=df[1]+40

sns.heatmap(df, cmap='viridis')
plt.show()

df_norm_col=(df-df.mean())/df.std()
sns.heatmap(df_norm_col, cmap='viridis')
plt.show()
```

```python
# 행 정규화
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 음수기호 표시
plt.rcParams['axes.unicode_minus'] = False

df = pd.DataFrame(np.random.randn(10,10)*4+3)
df.iloc[2]=df.iloc[2]+40

sns.heatmap(df, cmap='viridis')
plt.show()

df_norm_row = df.apply(lambda x: (x-x.mean())/x.std(), axis=1)

sns.heatmap(df_norm_row, cmap='viridis')
plt.show()
```

```python
# 상관계수&유클리드 거리
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

url = 'https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/mtcars.csv'
df = pd.read_csv(url)
df = df.set_index('model')

# 상관계수
sns.clustermap(df, metric="correlation",
               standard_scale=1)
plt.show()

# 유클리드 거리
sns.clustermap(df, metric="euclidean",
               standard_scale=1)
plt.show()
```

```python
# 클러스터 방법 설정
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

url = 'https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/mtcars.csv'
df = pd.read_csv(url)
df = df.set_index('model')

# 유클리드 거리
sns.clustermap(df, metric="euclidean",
               standard_scale=1,
               method="single")
plt.show()

sns.clustermap(df, metric="euclidean",
               standard_scale=1,
               method="ward")
plt.show()
```

```python
# 색상 변경
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

url = 'https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/mtcars.csv'
df = pd.read_csv(url)
df = df.set_index('model')

sns.clustermap(df, metric="euclidean",
               standard_scale=1,
               method="ward",
               cmap="mako")
plt.show()

sns.clustermap(df, metric="euclidean",
               standard_scale=1,
               method="ward",
               cmap="viridis")
plt.show()

sns.clustermap(df,
               metric="euclidean",
               standard_scale=1,
               method="ward",
               cmap="Blues")
plt.show()
```

```python
# 이상값 무시하고 맵표현: robust=True
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

url = 'https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/mtcars.csv'
df = pd.read_csv(url)
df = df.set_index('model')

# 이상치 설정
df.loc[15:16, 'drat'] = 1000

# True
sns.clustermap(df,
               robust=True)
plt.show()

# False
sns.clustermap(df,
               robust=False)
plt.show()
```
