```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kde
df = sns.load_dataset('iris')
```

```python
# shade  fill 
sns.set_style("white")

sns.kdeplot(x=df.sepal_width, 
            y=df.sepal_length)
plt.show()

# bandwidth 
sns.kdeplot(x=df.sepal_width, 
            y=df.sepal_length, 
            cmap="Reds", 
            fill=True, bw_adjust=.5)
plt.show()

# threshold 
sns.kdeplot(x=df.sepal_width, 
            y=df.sepal_length, 
            cmap="Blues", 
            fill=True, thresh=0)
plt.show()
```

```python
x = np.random.normal(size=50000)
y = x * 3 + np.random.normal(size=50000)
```

```python
# 2D histogram / bins  x ,  y 
plt.hist2d(x, y, bins=(50, 50), cmap=plt.cm.jet)
plt.show()

plt.hist2d(x, y, bins=(300, 300), cmap=plt.cm.Reds)
plt.show()

plt.hist2d(x, y, bins=(300, 30), cmap=plt.cm.BuPu)
plt.show()
# cbar  
plt.hist2d(x, y, bins=(50, 50), cmap=plt.cm.Greys)
plt.colorbar()
plt.show()
```

```python
x = np.random.normal(size=500)
y = x * 3 + np.random.normal(size=500)
```

```python
nbins=300
k = kde.gaussian_kde([x,y])
xi, yi = np.mgrid[x.min():x.max():nbins*1j, y.min():y.max():nbins*1j]
zi = k(np.vstack([xi.flatten(), yi.flatten()]))

plt.pcolormesh(xi, yi, zi.reshape(xi.shape), shading='auto')
plt.show()

plt.pcolormesh(xi, yi, zi.reshape(xi.shape), shading='auto', cmap=plt.cm.Greens_r)
plt.show()
```
