```python
# 등고선 차트
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

sns.set_style("white")

sns.kdeplot(x=df.sepal_width,
            y=df.sepal_length)
plt.show()

sns.kdeplot(x=df.sepal_width,
            y=df.sepal_length,
            cmap="Reds",            # 색상
            shade=True,             # 채우기
            bw_adjust=.5            # 굴곡
            )
plt.show()

sns.kdeplot(x=df.sepal_width,
            y=df.sepal_length,
            cmap="Blues",
            shade=True,
            thresh=0               # 음영된 영역에서 누적 밀도 함수를 표시하지 않음
            )
plt.show()
```

```python
# 2D 히스토그램
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(size=50000)
y = x * 3 + np.random.normal(size=50000)

plt.hist2d(x, y,
           bins=(50, 50),       # 픽셀 수
           cmap=plt.cm.jet      # 색상
           )
plt.show()

plt.hist2d(x, y,
           bins=(300, 300),       # 픽셀 수
           cmap=plt.cm.jet)
plt.show()

plt.hist2d(x, y,
           bins=(300, 30),       # 픽셀 수
           cmap=plt.cm.jet)
plt.show()
```

```python
# 사용가능한 팔레트
plt.hist2d(x, y,
           bins = (50, 50),
           cmap=plt.cm.Reds         # 레드
           )
plt.show()

plt.hist2d(x, y,
           bins = (50, 50),
           cmap=plt.cm.BuPu         # 블루퍼플
           )
plt.show()

plt.hist2d(x, y,
           bins = (50, 50),
           cmap=plt.cm.Greys         # 그레이
           )
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import kde

x = np.random.normal(size=500)
y = x * 3 + np.random.normal(size=500)

nbins=300
k = kde.gaussian_kde([x,y])

xi, yi = np.mgrid[x.min():x.max():nbins*1j, y.min():y.max():nbins*1j]
zi = k(np.vstack([xi.flatten(), yi.flatten()]))
 
plt.pcolormesh(xi, yi, zi.reshape(xi.shape), shading='auto')
plt.show()

plt.pcolormesh(xi, yi, zi.reshape(xi.shape), shading='auto', cmap=plt.cm.Greens_r)
plt.show()
```
