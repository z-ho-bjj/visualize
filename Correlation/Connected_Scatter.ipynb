```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'x_axis' : range(1,10),
    'y_axis':np.random.randn(9)*80+range(1,10)
})
```

```python
plt.plot('x_axis',
           'y_axis',
           data=df,
           linestyle='-',
           marker='o')
plt.show()
```

```python
df = pd.DataFrame({
    'x_axis' : range(1,101),
    'y_axis':np.random.randn(100)*15+range(1,101),
    'z' : (np.random.randn(100)*15 + range(1,101))*2
})
```

```python
plt.plot('x_axis',
         'y_axis',
         data=df,
         marker='o',
         color='mediumvioletred')
```

```python
# subplot / row,column,count 순으로 작성
# 아래의 예시는 1행 2열을 만들고 그 중 첫번째와 두번째를 각각 그림
plt.subplot(121)
plt.plot('x_axis',
         'y_axis',
         data=df,
         marker='o',
         alpha=.4)

plt.subplot(122)
plt.plot('x_axis',
         'y_axis',
         data=df,
         linestyle='none',
         marker='o',
         color='orange',
         alpha=.3)

plt.show()
```

```python
# subplot / row,column,count 순으로 작성
# 아래의 예시는 2행 1열을 만들었음
plt.subplot(211)
plt.plot('x_axis',
         'y_axis',
         data=df,
         marker='o',
         alpha=.4)

plt.subplot(212)
plt.plot('x_axis',
         'y_axis',
         data=df,
         linestyle='none',
         marker='o',
         color='orange',
         alpha=.3)

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, 
                         ncols=1, 
                         sharex=True, 
                         sharey=True)

axes[0].plot('x_axis',
             'y_axis',
             data=df,
             marker='o',
             alpha=.4)
axes[1].plot('x_axis',
             'y_axis',
             data=df,
             linestyle='none',
             marker='o',
             color='orange',
             alpha=.3)

plt.show()
```

```python
# subplots를 이용할 때, nrows와 ncols를 2이상의 수를 했다면,
# axes에 받아온 뒤 행과 열을 지정을 해주어야 한다
# ex) 위의 예시처럼 axes[0].plot을 하면 에러가 뜨게된다

fig, axes = plt.subplots(nrows=2, 
                         ncols=2, 
                         sharex=True, 
                         sharey=True)

axes[0,0].plot('x_axis','y_axis', data=df, marker='o', alpha=0.4)
axes[0,1].plot('x_axis','y_axis', data=df, linestyle='none', marker='o', color="orange", alpha=0.3)
axes[1,0].plot('x_axis','y_axis', data=df, linestyle='none', marker='D', color="green", alpha=0.3)
axes[1,1].plot('x_axis','y_axis', data=df, marker='o', color="grey", alpha=0.3)

# # 서브타이틀 공통이라면 아래의 코드로
# i=0
# for a in range(len(axes)): 
#     for b in range(len(axes[a])):
#         subplot_title=("Subplot"+str(i))
#         axes[a,b].set_title(subplot_title)
#         i=i+1

# 각각 달라야한다면,
axes[0,0].set_title('aaaa')
axes[0,1].set_title('b'*4)
axes[1,0].set_title('c'*4)
axes[1,1].set_title('d'*4)

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, 
                         ncols=2, 
                         sharex=True, 
                         sharey=True)

ax1 = plt.subplot2grid((2,2),(0,0),colspan=2)
ax1.plot('x_axis','y_axis', data=df, marker='o', alpha=0.4)
ax2 = plt.subplot2grid((2,2),(1,0),colspan=1)
ax2.plot('x_axis','y_axis', data=df, linestyle='none', marker='D', color="green", alpha=0.3)
ax3 = plt.subplot2grid((2,2),(1,1),colspan=1)
ax3.plot('x_axis','y_axis', data=df, marker='o', color="grey", alpha=0.3)

plt.show()
```
