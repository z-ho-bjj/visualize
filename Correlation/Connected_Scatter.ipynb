```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'x_axis' : range(1,10),
    'y_axis':np.random.randn(9)*80+range(1,10)
})
```

```python
plt.plot('x_axis',
           'y_axis',
           data=df,
           linestyle='-',
           marker='o')
plt.show()
```

```python
df = pd.DataFrame({
    'x_axis' : range(1,101),
    'y_axis':np.random.randn(100)*15+range(1,101),
    'z' : (np.random.randn(100)*15 + range(1,101))*2
})
```

```python
plt.plot('x_axis',
         'y_axis',
         data=df,
         marker='o',
         color='mediumvioletred')
```

```python
# subplot / row,column,count  
#   1 2       
plt.subplot(121)
plt.plot('x_axis',
         'y_axis',
         data=df,
         marker='o',
         alpha=.4)

plt.subplot(122)
plt.plot('x_axis',
         'y_axis',
         data=df,
         linestyle='none',
         marker='o',
         color='orange',
         alpha=.3)

plt.show()
```

```python
# subplot / row,column,count  
#   2 1 
plt.subplot(211)
plt.plot('x_axis',
         'y_axis',
         data=df,
         marker='o',
         alpha=.4)

plt.subplot(212)
plt.plot('x_axis',
         'y_axis',
         data=df,
         linestyle='none',
         marker='o',
         color='orange',
         alpha=.3)

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, 
                         ncols=1, 
                         sharex=True, 
                         sharey=True)

axes[0].plot('x_axis',
             'y_axis',
             data=df,
             marker='o',
             alpha=.4)
axes[1].plot('x_axis',
             'y_axis',
             data=df,
             linestyle='none',
             marker='o',
             color='orange',
             alpha=.3)

plt.show()
```

```python
# subplots  , nrows ncols 2  ,
# axes       
# ex)   axes[0].plot   

fig, axes = plt.subplots(nrows=2, 
                         ncols=2, 
                         sharex=True, 
                         sharey=True)

axes[0,0].plot('x_axis','y_axis', data=df, marker='o', alpha=0.4)
axes[0,1].plot('x_axis','y_axis', data=df, linestyle='none', marker='o', color="orange", alpha=0.3)
axes[1,0].plot('x_axis','y_axis', data=df, linestyle='none', marker='D', color="green", alpha=0.3)
axes[1,1].plot('x_axis','y_axis', data=df, marker='o', color="grey", alpha=0.3)

# #    
# i=0
# for a in range(len(axes)): 
#     for b in range(len(axes[a])):
#         subplot_title=("Subplot"+str(i))
#         axes[a,b].set_title(subplot_title)
#         i=i+1

#  ,
axes[0,0].set_title('aaaa')
axes[0,1].set_title('b'*4)
axes[1,0].set_title('c'*4)
axes[1,1].set_title('d'*4)

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, 
                         ncols=2, 
                         sharex=True, 
                         sharey=True)

ax1 = plt.subplot2grid((2,2),(0,0),colspan=2)
ax1.plot('x_axis','y_axis', data=df, marker='o', alpha=0.4)
ax2 = plt.subplot2grid((2,2),(1,0),colspan=1)
ax2.plot('x_axis','y_axis', data=df, linestyle='none', marker='D', color="green", alpha=0.3)
ax3 = plt.subplot2grid((2,2),(1,1),colspan=1)
ax3.plot('x_axis','y_axis', data=df, marker='o', color="grey", alpha=0.3)

plt.show()
```
