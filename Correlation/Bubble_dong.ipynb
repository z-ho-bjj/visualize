```python
# 거품크기
import matplotlib.pyplot as plt
import seaborn as sns
from gapminder import gapminder

data = gapminder.loc[gapminder.year == 2007]

sns.scatterplot(data=data,
                x="gdpPercap",          # x축
                y="lifeExp",            # y축
                size="pop",             # 버블 종류
                alpha=0.5,              # 투명도
                sizes=(20, 800)         # 버블크기
                )

plt.show()
```

```python
# 버블색상
import matplotlib.pyplot as plt
import seaborn as sns
from gapminder import gapminder

data = gapminder.loc[gapminder.year == 2007]

sns.scatterplot(data=data,
                x="gdpPercap",          # x축
                y="lifeExp",            # y축
                size="pop",             # 버블 종류
                hue="continent",        # 그룹
                alpha=0.5,              # 투명도
                sizes=(20, 800)         # 버블크기
                )

plt.show()
```

```python
# 기타 제어
sns.set_style("darkgrid")

sns.scatterplot(data=data,
                x="gdpPercap",          # x축
                y="lifeExp",            # y축
                size="pop",             # 버블 종류
                hue="continent",        # 그룹
                palette="viridis",      # 색상 표현
                alpha=0.5,              # 투명도
                sizes=(10, 1000)        # 버블크기
                )

plt.xlabel("Gdp per Capita")
plt.ylabel("Life Expectancy")

plt.show()
```

```python
# matplotlib: 기본 버블차트
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(40)
y = np.random.rand(40)
z = np.random.rand(40)

plt.scatter(x,
            y,
            s=z*1000,       # 버블 크기
            alpha=0.5       # 투명도
            )

plt.show()
```

```python
# 색상
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(40)
y = np.random.rand(40)
z = np.random.rand(40)

plt.scatter(x,
            y,
            s=z*1000,       # 버블 크기
            c="red",        # 버블 색상
            alpha=0.5       # 투명도
            )

plt.show()
```

```python
# 모양
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(40)
y = np.random.rand(40)
z = np.random.rand(40)

plt.scatter(x,
            y,
            s=z*1000,       # 버블 크기
            c="red",        # 버블 색상
            marker="D",     # 버블 모양
            alpha=0.5       # 투명도
            )

plt.show()
```

```python
# 가장자리 굵기
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(40)
y = np.random.rand(40)
z = np.random.rand(40)

plt.scatter(x,
            y,
            s=z*1000,       # 버블 크기
            c="green",      # 버블 색상
            marker="D",     # 버블 모양
            alpha=0.5,      # 투명도
            linewidths=6    # 가장자리 굵기
            )

plt.show()
```

```python
# seaborn 스타일 반영
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

x = np.random.rand(40)
y = np.random.rand(40)
z = np.random.rand(40)

# seaborn 스타일 반영
sns.set_theme(style="white")

plt.scatter(x,
            y,
            s=z*1000,       # 버블 크기
            c="green",      # 버블 색상
            marker="D",     # 버블 모양
            alpha=0.5,      # 투명도
            linewidths=6    # 가장자리 굵기
            )

plt.show()
```

```python
# 버블차트 4차원(색상) 표현
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

x = np.random.rand(15)
y = x+np.random.rand(15)
z = x+np.random.rand(15)
z = z*z

plt.scatter(x,                  # x 좌표값
            y,                  # y 좌표값
            s=z * 2000,         # 점 크기
            c=x,                # 색상 (x 값을 기준으로 색상 설정)
            cmap="Blues",       # 컬러맵 설정 (팔레트로 Blues 사용)
            alpha=0.4,          # 점의 투명도 (0에 가까울수록 투명, 1에 가까울수록 불투명)
            edgecolors="grey",  # 점의 테두리 색상 설정
            linewidth=2         # 점의 테두리 두께 설정
            )

plt.xlabel("the X axis")
plt.ylabel("the Y axis")
plt.title("A colored bubble plot")

plt.show()
```
