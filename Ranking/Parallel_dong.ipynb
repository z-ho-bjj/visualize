```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
import seaborn as sns

data = sns.load_dataset('iris')

parallel_coordinates(data,
                     'species',
                     colormap=plt.get_cmap("Set2")      # 색상표
                     )
plt.show()
```

```python
parallel_coordinates(data,
                     'species',
                     #colormap=plt.get_cmap("Set2")      # 색상표
                     color=('r','g','b'),                # 색상 지정 
                     alpha=0.5)
plt.show()
```
