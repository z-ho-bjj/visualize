```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
year = [2018,2019,2020,2021,2022]
ev = [55756, 89918, 134962, 231443, 389855]
charger = [27352, 44792, 64188, 106701, 194081]
```

```python
df = pd.DataFrame(data=[ev, charger], columns=year, index=['ev','charger'])
```

```python
fig, ax1 = plt.subplots()

ax1.bar(year, ev, color='skyblue', alpha=0.7, width=0.7)

ax2 = ax1.twinx()
ax2.plot(year, charger, '-s', color='green', markersize=7, linewidth=5, alpha=0.7)

plt.show()
```

```python
import seaborn as sns
from matplotlib import font_manager, rc
```

```python
font_path = "C:/Windows/Fonts/malgun.ttf"
font = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font)
```

```python
fig = plt.figure()
fig.set_facecolor('white')

 
colors = sns.color_palette('summer', len(year)) ##   
 
xtick_label_position = list(range(len(year))) ## x    x
plt.xticks(xtick_label_position, year) ## x   
 
bar = plt.bar(xtick_label_position, ev, color=colors) ##  
plt.plot(xtick_label_position, charger, color='b',
         linestyle='--', marker='o') ##   


current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values])

for rect in bar:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '{:,.0f}'.format(height), ha='center', va='bottom', size = 12)



plt.title('    ', fontsize=20)
plt.grid(visible=True, axis='y')
plt.show()
```

```python
fig = plt.figure()
fig.set_facecolor('white')
ax = fig.add_subplot()
colors = sns.color_palette('summer', len(year)) ##   
 
xtick_label_position = list(range(len(year))) ## x    x
plt.xticks(xtick_label_position, year) ## x   
 
bar = plt.bar(xtick_label_position, ev, color=colors) ##  
line_plot = plt.plot(xtick_label_position, charger, color='b', linestyle='--', marker='o') 
line_plot = line_plot[0] ##   

current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values])

for rect in bar:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '{:,.0f}'.format(height), ha='center', va='bottom', size = 12)

 
bbox = dict( ##    
    boxstyle='square', #  
    facecolor='white', #  
)
 
for coord in list(line_plot.get_xydata()):
    ax.text(coord[0]-.3,coord[1]-5000,'{:,.0f}'.format(coord[1]),fontsize=8, ha='center', bbox=bbox)


plt.title('    ', fontsize=20)
# plt.grid(visible=True, axis='y')
plt.show()
```
