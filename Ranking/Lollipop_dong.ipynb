```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.DataFrame({'group':list(map(chr, range(65, 85))), 'values':np.random.uniform(size=20) })

ordered_df = df.sort_values(by='values')        # 정렬
my_range=range(1,len(df.index)+1)

plt.stem(ordered_df['values'])
plt.xticks( my_range, ordered_df['group'])
plt.show()

# 줄기
plt.hlines(y=my_range,                        # 수평
           xmin=0,                            # 줄기 시작
           xmax=ordered_df['values'],         # 줄기 끝
           color='skyblue'                    # 줄기 색상
           )
# 마커
plt.plot(ordered_df['values'],
         my_range, "D"                        # 마커 모양
         )

plt.yticks(my_range, ordered_df['group'])
plt.show()
```

```python
# 마커 설정
import matplotlib.pyplot as plt
import numpy as np
 
values=np.random.uniform(size=40)

plt.stem(values,
         markerfmt=' '          # 마커 공백
         )
plt.show()

(markers, stemlines, baseline) = plt.stem(values)           # 마커, 줄기, 기준선 값을 반환
plt.setp(markers,
         marker='D',                    # 마커 모양
         markersize=10,                 # 마커 크기
         markeredgecolor="orange",      # 마커 테두리 색상
         markeredgewidth=2              # 마커 테두리 굵기
         )
plt.show()
```

```python
# 기준선 설정
import matplotlib.pyplot as plt
import numpy as np

values=np.random.uniform(size=100)
 
plt.stem(values,
         markerfmt=' ',         # 마커 없음
         bottom=0.5             # 시작점 값
         )
plt.show()

(markers, stemlines, baseline) = plt.stem(values)           # 마커, 줄기, 기준선 값을 반환
plt.setp(baseline,
         visible=False          # 기준선 숨김
         )
plt.show()
 
plt.stem(values,
         basefmt=" "            # 기준선 공백
         )
plt.show()
 
(markers, stemlines, baseline) = plt.stem(values)
plt.setp(baseline,
         linestyle="-",        # 기준선 스타일
         color="grey",         # 기준선 색상
         linewidth=6           # 기준선 굵기
         )
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
 
df = pd.DataFrame({'group':list(map(chr, range(65, 85))), 'values':np.random.uniform(size=20) })

ordered_df = df.sort_values(by='values')
my_range=range(1,len(df.index)+1)

plt.hlines(y=my_range,
           xmin=0,
           xmax=ordered_df['values'],
           color='skyblue'                  # 줄기 색상
           )
plt.plot(ordered_df['values'],
         my_range,
         "o"                                # 마크모양
         )

plt.yticks(my_range, ordered_df['group'])
plt.title("A vertical lolipop plot", loc='left')
plt.xlabel('Value of the variable')
plt.ylabel('Group')

plt.show()
```

```python
# 그룹 강조
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
df = pd.DataFrame({'group':list(map(chr, range(65, 85))), 'values':np.random.uniform(size=20) })

ordered_df = df.sort_values(by='values')
my_range=range(1,len(df.index)+1)

# 그룹 B에 대한 강조 설정
my_color=np.where(ordered_df ['group']=='B', 'orange', 'skyblue')
my_size=np.where(ordered_df ['group']=='B', 70, 30)

plt.hlines(y=my_range, xmin=0, xmax=ordered_df['values'], color=my_color, alpha=0.4)
plt.scatter(ordered_df['values'], my_range, color=my_color, s=my_size, alpha=1)

plt.yticks(my_range, ordered_df['group'])
plt.title("What about the B group?", loc='left')
plt.xlabel('Value of the variable')
plt.ylabel('Group')

plt.show()
```

```python
# 2개 그룹 롤리팝 차트
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
value1=np.random.uniform(size=20)
value2=value1+np.random.uniform(size=20)/4
df = pd.DataFrame({'group':list(map(chr, range(65, 85))), 'value1':value1 , 'value2':value2 })

ordered_df = df.sort_values(by='value1')
my_range=range(1,len(df.index)+1)

plt.hlines(y=my_range,
           xmin=ordered_df['value1'],           # 그룹1
           xmax=ordered_df['value2'],           # 그룹2
           color='grey',                        # 줄기 색상
           alpha=0.4                            # 투명도
           )
plt.scatter(ordered_df['value1'],               # 그룹1 마크
            my_range,
            color='skyblue',
            alpha=1,
            label='value1'
            )
plt.scatter(ordered_df['value2'],               # 그룹2 마크
            my_range,
            color='green',
            alpha=0.4 ,
            label='value2'
            )

plt.legend()            # 범례표시

plt.yticks(my_range, ordered_df['group'])
plt.title("Comparison of the value 1 and the value 2", loc='left')
plt.xlabel('Value of the variables')
plt.ylabel('Group')

plt.show()
```

```python
# 조건부 색상 설정
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x) + np.random.uniform(size=len(x)) - 0.2

# 조건부 설정
my_color = np.where(y>=0, 'orange', 'skyblue')

plt.rcParams['axes.unicode_minus'] = False

plt.vlines(x=x, ymin=0, ymax=y, color=my_color, alpha=0.4)
plt.scatter(x, y, color=my_color, s=1, alpha=1)

plt.title("Evolution of the value of ...", loc='left')
plt.xlabel('Value of the variable')
plt.ylabel('Group')

plt.show()
```
