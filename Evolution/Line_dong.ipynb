```python
# 라인색상 지정
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df=pd.DataFrame({'x_values': range(1,11), 'y_values': np.random.randn(10) })

plt.plot('x_values',
         'y_values',
         data=df,
         color='skyblue'        # 라인색 지정
         )
plt.show()

plt.plot('x_values',
         'y_values',
         data=df,
         color='skyblue',       # 라인색 지정
         alpha=0.3              # 투명도
         )
plt.show()
```

```python
# 시계열 차트
import matplotlib.pyplot as plt
import pandas as pd

# 2013~2019 사이의 비트코인 가격의 변화
data = pd.read_csv(
    "https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/3_TwoNumOrdered.csv", 
    delim_whitespace=True
)

data["date"] = pd.to_datetime(data["date"])         # datetime형식으로 변환

date = data["date"]
value = data["value"]

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(date, value)
```

```python
# X축 날짜 설정
import matplotlib.dates as mdates           # 연도가 아닌 다른 항목에 대해 가로축을 설정할때 유용

fig, ax = plt.subplots(figsize=(8, 6))

half_year_locator = mdates.MonthLocator(interval=6)            # 눈금에 표할 개월 수
ax.xaxis.set_major_locator(half_year_locator)

ax.plot(date, value)

plt.show()
```

```python
import matplotlib.dates as mdates           # 연도가 아닌 다른 항목에 대해 가로축을 설정할때 유용

fig, ax = plt.subplots(figsize=(8, 6))

half_year_locator = mdates.MonthLocator(interval=6)            # 눈금에 표할 개월 수
year_month_formatter = mdates.DateFormatter("%Y-%m")           # 연도4, 월2 자리로 표시

ax.xaxis.set_major_locator(half_year_locator)
ax.xaxis.set_major_formatter(year_month_formatter)

ax.plot(date, value)
plt.show()
```

```python
# 레이블 자동회전
fig, ax = plt.subplots(figsize=(8, 6))

ax.xaxis.set_major_locator(half_year_locator)
ax.xaxis.set_major_formatter(year_month_formatter)
ax.plot(date, value)

fig.autofmt_xdate()         # 레이블 자동 회전

plt.show()
```
